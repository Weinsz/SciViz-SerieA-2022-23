```python
import ScraperFC as sfc
import traceback

# Initialize the Understat scraper
scraper = sfc.FBRef()
try:
    # Scrape the table
    standard_stats = scraper.scrape_stats(year=2023, league="Serie A", stat_category="standard")
    #formations = scraper.scrape_formations(year=2023, league="Serie A")

except:
    # Catch and print any exceptions. This allows us to still close the
    # scraper below, even if an exception occurs.
    traceback.print_exc()
finally:
    # It"s important to close the scraper when you"re done with it. Otherwise,
    # you"ll have a bunch of webdrivers open and running in the background.
    scraper.close()

standard_stats
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df_completo = standard_stats[2][[('Unnamed: 1_level_0', 'Player'), ('Per 90 Minutes', 'npxG'), ('Per 90 Minutes', 'G-PK')]]
df_dati =  standard_stats[2][[('Per 90 Minutes', 'npxG'), ('Per 90 Minutes', 'G-PK')]]['Per 90 Minutes']
df_dati['G-PK'] = pd.to_numeric(df_dati['G-PK'])
df_dati['npxG'] = pd.to_numeric(df_dati['npxG'])
plt.scatter(df_dati['npxG'], df_dati['G-PK'])
plt.show()
```
